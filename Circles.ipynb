```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def circleExact(x, y, r, fill=0):
    RGB = np.zeros((200, 200, 3), dtype = np.uint8)
    RGB.fill(255)
    dr = r
    xmax = min(200, x+r+1)
    xmin = max(0, x-r)
    ymax = min(200, y+r+1)
    ymin = max(0, y-r)
    for i in range(ymin, ymax):
        for j in range (xmin, xmax):    
            xx = j - x
            yy = i - y
            if xx*xx +yy*yy <= r*r+r and xx*xx+yy*yy >= r*r-r:
                RGB[i][j] = 0
    return RGB
```

```python
data = circleExact(120, 120, 70)
plt.imshow(data, interpolation='none', aspect=1)
plt.show()
```

```python
def circleFill(x, y, r):
    RGB = np.zeros((200, 200, 3), dtype = np.uint8)
    RGB.fill(255)
    dr = r
    xmax = min(200, x+r+1)
    xmin = max(0, x-r)
    ymax = min(200, y+r+1)
    ymin = max(0, y-r)
    for i in range(ymin, ymax):
        for j in range (xmin, xmax):    
            xx = j - x
            yy = i - y
            if xx*xx +yy*yy <= r*r+r:
                RGB[i][j] = 0
    return RGB
```

```python
data = circleFill(120, 120, 70)
plt.imshow(data, interpolation='none', aspect=1)
plt.show()
```

```python
def circleF(xc, yc, r, fill = 1):
    if fill == 1:
        data = circleFill(xc, yc, r)
    elif fill == 0:
        data = circleExact(xc, yc, r)
        
    return data
```

```python
data = circleF(120, 120, 70, 1)
plt.imshow(data, interpolation='none', aspect=1)
plt.show()
```

```python
data = circleF(120, 120, 70, 0)
plt.imshow(data, interpolation='none', aspect=1)
plt.show()
```

```python
def circleB(xc, yc, r):
    RGB = np.zeros((200, 200, 3), dtype = np.uint8)
    RGB.fill(255)
    d = 3-2*r
    y=r
    i=0
    while i<=y:
        y1 = min(yc + y, 200)
        y2 = max(yc - y, 0)
        y3 = min(yc + i, 200)
        y4 = max(yc - i, 0)
        
        x1 = min(xc + i, 200)
        x2 = max(xc - i, 0)
        x3 = max(xc - y, 0)
        x4 = min(xc + y, 200)
        
        RGB[y1, x1] = 0.0
        RGB[y2, x1] = 0.0
        RGB[y1, x2] = 0.0
        RGB[y2, x2] = 0.0
        RGB[y3, x3] = 0.0
        RGB[y4, x3] = 0.0
        RGB[y3, x4] = 0.0
        RGB[y4, x4] = 0.0
        
        i+= 1
        if (d<=0):
            d+= i*4+6
        else:
            d+= 4*(i-y)+10
            y-=1
    return RGB
```

```python
data = circleB(50, 120, 70)
plt.imshow(data, interpolation='none', aspect=1)
plt.show()
```
