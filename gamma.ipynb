```python
from IPython.display import display
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
path="./Koloseum.png"
img = Image.open(path)

display(img)
```

```python

RGB = np.array(img) # tworzymy tablic numpy RGB na podstawie rysunku. Uwaga! wartosci sa od 0 do 255 powstaje tablica typu numpy.uint8t

print(len(RGB))
print(len(RGB[0]))
```

```python
RGB2= np.zeros((len(RGB), len(RGB[0]),3), dtype=np.uint8,) # tworzymy nowa tablice typu numpy.uint8t

img2 = Image.fromarray(RGB2, 'RGB') # tworzymy nowy rysunek na podstawie tablicy RGB2 o rozmiarach takich samych jak img

display(img2) #wywietlamy obrazek
```

```python
def dodaj(data, value):
    for i in range(len(data)):
        for j in range(len(data[i])):
            for z in range(3):
                pomoc = np.int16(data[i][j][z])+value
                if pomoc > 255:
                    pomoc = 255
                elif pomoc < 0:
                    pomoc = 0
                data[i][j][z] = pomoc
    img3 = Image.fromarray(data, 'RGB')
    display(img3)
```

```python
dodaj(RGB, -40)
```

```python
def gammaK(data, gamma):
    #funkcja dokonuje korekcji gamma wskazanego rysubnku, prosze przetestowac jej dzialanie i narysowac rysunki 
    # uzyskane dla gamma = 0.5, 0.75, 1., 1.25, 1.5, 2., 4.
    for i in range(len(data)):
        for j in range(len(data[i])):
            for z in range(3):
                pomoc=data[i][j][z]
                data[i][j][z]=255*(pomoc/255.)**(1/gamma)
    img4 = Image.fromarray(data, 'RGB')
    display(img4)
```

```python
gammaK(RGB, 4)
```
